```python
# Embedding model name: nomic-embed-text:latest
```

```python
from langchain_ollama import OllamaEmbeddings
```

```python
embedding = OllamaEmbeddings(
    model="nomic-embed-text:latest"
)
```

```python
vectorised_res = embedding.embed_documents(["This is one string.",  "This is another string."])
```

```python
vectorised_res
```

```python
len(vectorised_res)
```

```python
from langchain_community.vectorstores import Chroma
```

```python
db = Chroma.from_texts(["This is one string.",  "This is another string."], embedding)
```

```python
res = db.search("?", search_type='similarity')
res
```
